```python
import numpy as np
import matplotlib.pyplot as plt
```

两条曲线再一个图上

```python
X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
def plot_slope(X, Y):
 Xs = X[1:] - X[:-1]
 Ys = Y[1:] - Y[:-1]
 plt.plot(X[1:], Ys / Xs)
X = np.linspace(-3, 3, 100)
Y = np.exp(-X ** 2)
plt.plot(X, Y)
plot_slope(X, Y)
plt.show()
```

从numpy多维数组中划线

```python
data = np.array([[0, 0, 6],
[1, 1, 5],
[2, 4, 4],
[4, 16, 3],
[5, 25, 2],
[6, 36, 1]])
for column in data.T:
 plt.plot(data[:,0], column)
plt.show()
```
