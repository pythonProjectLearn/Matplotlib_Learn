````python
# 查修改路径
import matplotlib
matplotlib.matplotlib_fname()
````

1. 先将windwos中fonts目录下的 simhei.ttf 文件 拷贝到/home/zt/anaconda3/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf目录中，
2. 删除~/.cache/matplotlib的缓冲目录,如果不删除缓存,后面就算修改了也无法显示中文
3. 修改配置文件：
```
vim /home/zt/anaconda3/lib/python3.6/site-packages/matplotlib/mpl-data/matplotlibrc
```

font.family         : sans-serif  # 将font.family之前的#号去掉  
font.sans-serif     : SimHei, Bitstream Vera Sans, Lucida Grande, Verdana, Geneva, Lucid, Arial, Helvetica, Avant Garde, sans-serif # 将font.sans-serif之前的#去掉,然后用SimHei 替换第一个字体的位置   
axes.unicode_minus  Fasle  # 将True改为False，作用就是解决负号'-'显示为方块的问题


经过上述修改之后,就可以显示中文了

````python
import matplotlib.pyplot as plt
plt.plot([-1,2,-5,3])  
plt.title(u'中文')  
plt.show()
````
