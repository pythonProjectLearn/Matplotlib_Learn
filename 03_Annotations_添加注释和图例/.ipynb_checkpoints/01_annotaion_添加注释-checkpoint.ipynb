# 添加注释的地方有
 
- Adding a title   标题
- Using LaTeX-style notations   公式
- Adding a label to each axis   坐标的label
- Adding text                    文字
- Adding arrows                  行          
- Adding a legend                标签
- Adding a grid                  网格
- Adding lines                   线条
- Adding shapes                  类型  
- Controlling tick labeling       
- Controlling tick labeling

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 01 plt.title 添加标题

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
plt.title('A polynomial')
plt.plot(X, Y, c = 'k')
plt.show()
```

### 02 在title上添加latex公式

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
plt.title('$f(x)=\\frac{1}{4}(x+4)(x+1)(x-2)$')
plt.plot(X, Y, c = 'k')
plt.show()
```

### 03 为每个坐标添加标签

plt.xlab()
plt.ylab()

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
plt.title('Power curve for airfoil KV873')
plt.xlabel('Air speed')
plt.ylabel('Total drag')
plt.plot(X, Y, c = 'k')
plt.show()
```

### 04 在图上添加文字 

plt.text()


```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
plt.text(-0.5, -0.25, 'Brackmard minimum')
plt.plot(X, Y, c = 'k')
plt.show()
```

### 04 方框bbox位置的控制

相对于文本的box中心点的四个方位，例如点(-0.5, -0.20)的方位
- 'center'
- 'top'
- 'bottom'
- 'baseline'
- 'left':
- 'right'


box的类型
- 'facecolor' : '.75',   控制box的颜色
- 'edgecolor' : 'k',     box的边缘颜色
- 'boxstyle' : 'round'   box的类型 还有’square‘
- 'alpha'  透明程度


```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
box = {
 'facecolor' : '.75',  # 控制box的颜色
 'edgecolor' : 'k',    # box的边缘颜色
 'boxstyle' : 'round'  # box的类型 还有’square‘
}
plt.text(-0.5, -0.20, 'Brackmard minimum', bbox = box)
plt.plot(X, Y, c='k')
plt.show()
```

### 05 添加箭头
plt.annotate() 构建了一个箭头组合

arrowprops参数中的字段：
- 'arrowstyle':'<-', '<', '-', 'wedge','simple'，'fancy'  箭头类型
- 'facecolor':箭头的形状
- 'edgecolor':箭头的边缘
- 'alpha':透明度

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
plt.annotate('Brackmard minimum',  # 箭头的标注
             ha = 'center',        # 箭头
             va = 'bottom',        # 箭尾部
             xytext = (-1.5, 3.),  # 箭头标注的文本的位置
             xy = (0.75, -2.7),    # 箭头指向的位置
             arrowprops = { 'facecolor' : 'black', 'shrink' : 0.05 }  # 箭头的类型
            )
plt.plot(X, Y)
plt.show()
```

### 06 添加说明标签 

首先是要在plt.plot()中给予label参数

plt.legend()的参数
-'loc': 最好是使用默认值，其他的方位有 'upper left', 'lower left','lower right', 'right', 'center left', 'center right', 'lowercenter', 'upper center', and 'center'.

- 'shadow':  True or False, 
- 'fancybox':  True or False 
- 'title': This renders the legend with the title passed as a parameter.
- 'ncol': 数字
  

```python
X = np.linspace(0, 6, 1024)
Y1 = np.sin(X)
Y2 = np.cos(X)
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X, Y1, c = 'k', lw = 3., label = 'sin(X)')
plt.plot(X, Y2, c = '.5', lw = 3., ls = '--', label = 'cos(X)')
plt.legend()
plt.show()
```

### 07 在 图上添加网格

plt.grid()

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
plt.plot(X, Y, c = 'k')
plt.grid(True, lw = 2, ls = '--', c = '.75')
plt.show()
```

### 08 添加线条 
plt.gca() 表示创建一个灰图

.add_line() 在灰图上添加线条

plt.Line2D()创建2D线条对象，在这个对象上，参数linestyle, linewidth, marker, markersize, color,都可以使用

plt.axis('scaled') 是使不同的数据被归一化在同一坐标线

```python
N = 16
for i in range(N):
 plt.gca().add_line(plt.Line2D((0, i), (N - i, 0), color = '.75'))
plt.grid(True)
plt.axis('scaled')
plt.show()
```

### 09 创建不同形状的物体 

不同的形状在import matplotlib.patches as patches 对象中

渲染图
plt.gca().add_patch(shape) 用这个shape形状对象去渲染这个灰度图

boxstyle的备选项还有：'larrow', 'rarrow', 'round', 'round4', 'roundtooth', 'sawtooth',
or 'square')

```python
import matplotlib.patches as patches
# Circle 
shape = patches.Circle((0, 0), radius = 1., color = '.75')
plt.gca().add_patch(shape)

# Rectangle
shape = patches.Rectangle((2.5, -.5), 2., 1., color = '.75')
plt.gca().add_patch(shape)

# Ellipse
shape = patches.Ellipse((0, -2.), 2., 1., angle = 45., color =
 '.75')
plt.gca().add_patch(shape)

# Fancy box
shape = patches.FancyBboxPatch((2.5, -2.5), 2., 1., boxstyle =
 'sawtooth', color = '.75')
plt.gca().add_patch(shape)

# Display all
plt.grid(True)
plt.axis('scaled')
plt.show()
```
