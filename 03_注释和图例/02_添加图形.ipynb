## 添加图形，用到matplotlib.patches对象
所有形状都在这个对象里

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import numpy as np
```

### 01 添加形状

plt.gca().add_patch(shape)中的add_patch()是将图形对象，渲染到gca()灰度图对象当中

```python
# Circle
shape = patches.Circle((0, 0), radius = 1., color = '.75')
plt.gca().add_patch(shape)
# Rectangle 矩形的 左下角， 长 宽
shape = patches.Rectangle((2.5, -.5), 2., 1., color = '.75')
plt.gca().add_patch(shape)
# Ellipse
shape = patches.Ellipse((0, -2.), 2., 1., angle = 45., color =
 '.75')
plt.gca().add_patch(shape)
# Fancy box
shape = patches.FancyBboxPatch((2.5, -2.5), 2., 1., boxstyle =
 'sawtooth', color = '.75')
plt.gca().add_patch(shape)
# Display all
plt.grid(True)
plt.axis('scaled')
plt.show()
```

### 02 添加 多边形

matplotlib.patches.Polygon() 接收的是一个array-2d，组成一个闭合的回路


```python
theta = np.linspace(0, 2 * np.pi, 8)
points = np.vstack((np.cos(theta), np.sin(theta))).transpose()
plt.gca().add_patch(patches.Polygon(points, color = '.75'))
plt.grid(True)
plt.axis('scaled')
plt.show()
```

### 03 在图上添加线条

- 首先是渲染上 图形
- 再渲染上多边形

```python
theta = np.linspace(0, 2 * np.pi, 6)

points = np.vstack((np.cos(theta), np.sin(theta))).transpose()
plt.gca().add_patch(plt.Circle((0, 0), radius = 1., color =
 '.75'))

plt.gca().add_patch(plt.Polygon(points, closed=None, fill=None,
 lw = 3., ls = 'dashed', edgecolor = 'k'))
plt.grid(True)
plt.axis('scaled')
plt.show()
```

```python
radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)

plt.subplot(111, projection='polar')
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, 0.5*np.cos(20*theta), "b-")
plt.show()
```
