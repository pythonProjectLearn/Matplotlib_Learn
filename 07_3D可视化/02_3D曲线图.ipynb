```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
a, b, c = 10., 28., 8. / 3.
def lorenz_map(X, dt = 1e-2):
    X_dt = np.array([a * (X[1] - X[0]),
    X[0] * (b - X[2]) - X[1],
    X[0] * X[1] - c * X[2]])
    return X + dt * X_dt
points = np.zeros((10000, 3))
X = np.array([.1, .0, .0])
for i in range(points.shape[0]):
    points[i], X = X, lorenz_map(X)
    
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot(points[:, 0], points[:, 1], points[:, 2], c = 'k')
plt.show()
```

```python
from matplotlib import cm
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
X, Y = np.meshgrid(x, y)
Z = np.sinc(np.sqrt(X ** 2 + Y ** 2))
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z, cmap=cm.gray)
plt.show()
```

可以控制的一些参数

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z,
                cmap=cm.gray,
                linewidth=0,
                antialiased=False)
plt.show()
```

颜色color = 'w'

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z,
#                 cmap=cm.gray,
                color = 'w',
                linewidth=0,
                antialiased=False)
plt.show()
```

ax.plot_wireframe网格图

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_wireframe(X, Y, Z, cstride=8, rstride=8, color = 'k')
plt.show()
```
