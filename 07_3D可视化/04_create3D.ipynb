```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X ** 2 + Y ** 2))
u = np.exp(-(x ** 2))
```

ax.plot()中的参数
    - zdir是将2D的数据映射到X,Y,Z的哪个轴上
    - zs 确定平面的偏移

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.set_zlim3d(0, 3)
ax.plot(x, u, zs=3, zdir='y', lw = 2, color = '.75')
ax.plot(x, u, zs=-3, zdir='x', lw = 2., color = 'k')
ax.plot_surface(X, Y, Z, color = 'w')
plt.show()
```

```python
# Data generation
alpha = 1. / np.linspace(1, 8, 5)
t = np.linspace(0, 5, 16)
T, A = np.meshgrid(t, alpha)
data = np.exp(-T * A)
```

```python
from matplotlib import cm
import matplotlib.colors as col

# Plotting
fig = plt.figure()
ax = fig.gca(projection = '3d')
# 创建新的调色板
cmap = cm.ScalarMappable(col.Normalize(0, len(alpha)), cm.gray)

# 在三维图上，依次画出每根2D的柱子
for i, row in enumerate(data):
    ax.bar(4 * t, row, zs=i, zdir='y', alpha=0.8, color=cmap.to_rgba(i))
plt.show()
```

```python
# Data generation
alpha = np.linspace(1, 8, 5)
t = np.linspace(0, 5, 16)

T, A = np.meshgrid(t, alpha)
data = np.exp(-T * (1. / A))

# 建立每个柱状图的坐标,依次排列，成1维
Xi = T.flatten()
Yi = A.flatten()
Zi = np.zeros(data.size)

dx = .25 * np.ones(data.size)
dy = .25 * np.ones(data.size)
dz = data.flatten()
```

```python
Xi
```

创建的柱子是3D，Xi Yi Zi 为对应柱子的坐标  dx dy dz 反应柱子的长宽高

```python
# Plotting
fig = plt.figure()
ax = fig.gca(projection = '3d')

ax.set_xlabel('T')
ax.set_ylabel('Alpha')
ax.bar3d(Xi, Yi, Zi, dx, dy, dz, color = 'w')
plt.show()
```
